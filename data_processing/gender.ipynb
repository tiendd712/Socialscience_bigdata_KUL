```python
import pandas as pd
import numpy as np
from deepface import DeepFace
import cv2
from urllib.request import urlopen
import matplotlib.pyplot as plt
import numpy as np
from io import BytesIO
import requests
import json
```

```python
def url_to_image(url, readFlag=cv2.IMREAD_COLOR):
    # download the image, convert it to a NumPy array, and then read
    # it into OpenCV format
    resp = urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, readFlag)

    # return the image
    return image
```

```python
def gender_predict(employee_id):
    image_url = image_data.loc[image_data.employee_id == employee_id, "url_400_400"].values[0]
    try:
        img = url_to_image(image_url)
        prediction = DeepFace.analyze(img, actions=['gender'], enforce_detection = False)
        dict_pred = {employee_id : prediction}
        return(dict_pred)
    except Exception as e:
        dict_pred = {employee_id : "NA"}
        return(dict_pred)
   
```

```python

file_path = '../Socialscience_bigdata_KUL/data_processing/gender_predic_dict.json'

try:
    with open(file_path, 'r') as file:
        gender_predic_dict = json.load(file)
except json.JSONDecodeError as e:
    print(f"Error decoding JSON file: {e}")
```

```python
gender_predict = []
employee_id = []
for key in gender_predic_dict:
    if gender_predic_dict[key] == "NA":
        employee_id += [key]
        gender_predict += ["NA"]
    if len(gender_predic_dict[key]) == 1:
        employee_id += [key]
        gender_predict += [gender_predic_dict[key][0]["dominant_gender"]]
    if len(gender_predic_dict[key]) > 1 and gender_predic_dict[key] != "NA":
        employee_id += [key]
        gender_vec  = []
        for i in range(len(gender_predic_dict[key])):
            gender_vec += [gender_predic_dict[key][i]["dominant_gender"]]
        gender_vec = sorted(list(set(gender_vec)))
        gender_predict += [', '.join(value for value in gender_vec)]


```

```python
gender_predict_data = pd.DataFrame({"employee_id" : employee_id,
                                    "gender_predict": gender_predict})
```

```python
image_df = pd.read_csv('image_data.csv')
gender_df = pd.read_csv('gender_predict_data.csv')
gender_df_full = gender_df.merge(image_df, how='left', on='employee_id')
gender_df_full['gender_predict'] = gender_df_full['gender_predict'].astype("str")
gender_df_full['predictions'] = gender_df_full['gender_predict'].apply(lambda x: 2 if ',' in x else 1) # sort values so the one with 2 labels are ranked first
gender_df_full.sort_values(by="predictions", ascending=False, inplace=True)
gender_df_full.to_csv("gender_prediction_and_image.csv")
```

```python
from matplotlib.backends.backend_pdf import PdfPages

def save_image(filename):
    p = PdfPages(filename)
    for index, row in gender_df_full.iterrows():
        label = row['gender_predict']
        image_url = row['url_400_400']
        image_name = row['employee_id']
        
        try:
            image_array = url_to_image(image_url)
        except Exception as e:
            print(f"Error occurred for image {image_name}: {e}")
            continue
        
        fig = plt.figure()
        plt.imshow(image_array[:,:,::-1])
        plt.title('image ' + str(image_name) + ' label:' + (label))
        fig.savefig(p, format='pdf') 
    
    # close the object
    p.close()
```

```python
# name Pdf file
filename = "gender_image.pdf"  
save_image(filename)  
```
