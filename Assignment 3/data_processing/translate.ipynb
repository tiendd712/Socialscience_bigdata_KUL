```python
import pandas as pd
import numpy as np
from langdetect import detect
from googletrans import Translator
import time     
```

## Translate experience data

```python
experience_data = pd.read_csv("C:/Users/doduc/Github/Socialscience_bigdata_KUL/data_processing/experience_data.csv")
```

```python
experience_data.head(5)
```

```python
experience_data["title_trans"] = "1"
```

```python
experience_data.head(5)
```

```python
translator = Translator()
```

```python
for i in range(0,len(experience_data["title"])):
  if experience_data.iloc[i, 13] == "1" :
    try:
      translation = translator.translate(experience_data.iloc[i, 2], dest='en').text
      experience_data.iloc[i, 13] = translation
      

    except Exception as e:
      print(f"Error occurred: {str(e)}")
      print("Continuing after pause...")
      time.sleep(5)
      continue
```
